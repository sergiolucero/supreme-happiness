# Reclustering

```python
import pickle

modelos = pickle.load(open('modelos.pk', 'rb'))
#modelos = {'kmeans': (kmeans_model, labels, clusters),
#            'pca': (pca, datapoint), 'd2v': d2v_model}

kmeans_model, labels, clusters = modelos['kmeans']
pca, datapoint = modelos['pca']
d2v_model = modelos['d2v']
```

```python
import matplotlib.pyplot as plt                  #Elbow Method
from sklearn.cluster import KMeans

nc = range(1,10)

kmeans = []
for i in nc:
    kmeans.append(KMeans(n_clusters = i, n_init = 100, max_iter = 500, precompute_distances = 'auto' ))
    
score = [kmeans[i].fit(d2v_model.docvecs.doctag_syn0).score(d2v_model.docvecs.doctag_syn0) for i in range(len(kmeans))]

fig, ax = plt.subplots(1, figsize=(20,12))
plt.plot(nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```
